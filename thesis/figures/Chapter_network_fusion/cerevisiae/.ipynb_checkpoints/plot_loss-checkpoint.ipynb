# Plot loss for thesis

```python
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import glob
```

```python
def plotter(ax, d, c="k", label=None):
    ax.plot(
        np.arange(1, len(d['val_loss']) + 1),
        np.array(d['val_loss']),
        color=c,
        label=label,
        )

    ax.plot(
        np.arange(1, len(d['loss']) + 1),
        np.array(d['loss']),
        linestyle='--', alpha=0.5,
        color=c,
        )

    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    
    if label:
        ax.legend()
        
def plot_loss(path, fname=None):
    with open(path, "rb") as f:
        d = pickle.load(f)

    fig, ax = plt.subplots(1, figsize=(3, 2))
    plotter(ax, d)
    fig.tight_layout()
    if fname:
        fig.savefig(fname)
        
def plot_multiple_losses(path, name_structure, fname=None, *, major_locator=.2, minor_locator=.05):
    fig, ax = plt.subplots(1, figsize=(3, 2))
    cs = sns.husl_palette(len(glob.glob(path+"/*/")))

    for i, arch in enumerate(glob.glob(path+"/*/")):
        arch = arch.split("/")[-2]

        with open(os.path.join(path, arch, name_structure.format(arch)), "rb") as f:
            d = pickle.load(f)

            ax.plot(
                np.arange(1, len(d['val_loss']) + 1),
                np.array(d['val_loss']),
                color=cs[i],
                label=arch,
                alpha=.75,
                )

    ax.set_yscale("log", nonposy='clip')
    ax.yaxis.set_major_locator(mpl.ticker.MultipleLocator(major_locator))
    ax.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(minor_locator))
    ax.yaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
    ax.yaxis.set_minor_formatter(mpl.ticker.NullFormatter())

    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    ax.legend(loc="upper left", bbox_to_anchor=(1,1), frameon=True, ncol=2)
    fig.tight_layout()
    
    if fname:
        fig.savefig(fname, bbox_inches="tight")
```

Reproduce deepNF

```python
path = os.path.join(os.path.expandvars("$CEREVISIAEDATA"), "deepNF", "models", "180807_77cd62b", "cerevisiae_MDA_arch_2000-600_training_history.pkl")
plot_loss(path, "deepNF/180807_77cd62b/cerevisiae_MDA_arch_2000-600.pdf")
```
