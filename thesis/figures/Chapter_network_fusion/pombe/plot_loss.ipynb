# Plot loss for thesis

```python
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import glob
```

6 networks

```python
path = os.path.join(os.path.expandvars("$AGAPEDATA"), "deepNF", "models", "180820_45c86de")
```

```python
fig, ax = plt.subplots(1, figsize=(3, 2))
cs = sns.husl_palette(len(glob.glob(path+"/*/")))

for i, arch in enumerate(glob.glob(path+"/*/")):
    arch = arch.split("/")[-2]
    
    with open(os.path.join(path, arch, f"yeast_MDA_arch_{arch}_training_history.pkl"), "rb") as f:
        d = pickle.load(f)
        
        ax.plot(
            np.arange(1, len(d['val_loss']) + 1),
            np.array(d['val_loss']),
            color=cs[i],
            label=arch,
            alpha=.75,
            )

ax.set_yscale("log", nonposy='clip')
ax.yaxis.set_major_locator(mpl.ticker.MultipleLocator(.1))
ax.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(.02))
ax.yaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.yaxis.set_minor_formatter(mpl.ticker.NullFormatter())

ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.legend(loc="upper left", bbox_to_anchor=(1,1), frameon=True, ncol=2)
fig.tight_layout()

fig.savefig("autoencoder/architectures_6-networks.pdf", bbox_inches="tight")
```
